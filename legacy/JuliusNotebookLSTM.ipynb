```python
all_labels = set([tuple([i[0]-65, int(chr(i[1])+chr(i[2])), i[3]-65]) for i in np.load("label_pop.npy")])
```

```python
import random
def convert_label_to_vector(label):
    first_half, second_half = label.split(" ")
    return tuple([ord(first_half[0])-65, int(first_half[1:3]), ord(first_half[3])-65])

def data_generator():
    n = 0
    for label, vector in zip(training_arr_labels, training_arr):
        label_vectors = [convert_label_to_vector(z.strip().strip('\'\"')) for z in label[0].split(',')]
        non_true_vectors = []
        while len(non_true_vectors) <= len(label_vectors):
            cur = random.sample(all_labels, 1)
            if cur in label_vectors:
                continue
            non_true_vectors.append(cur[0])
        left = np.vstack((np.vstack([label_dict[i] for i in label_vectors]), np.vstack([label_dict[i] for i in non_true_vectors])))
        labels = [1 for _ in label_vectors]+[0 for _ in non_true_vectors]
        right = np.vstack([vector]*len(label_vectors+non_true_vectors))
        for l, r, label in zip(left, right, labels):
            ret = (np.hstack((l, r)).reshape((1, 2000)), np.array([label]).reshape((1)))
            yield ret
def data_generator_test():
    n = 0
    for label, vector in zip(testing_arr_labels, testing_arr):
        label_vectors = [convert_label_to_vector(z.strip().strip('\'\"')) for z in label[0].split(',')]
        non_true_vectors = []
        while len(non_true_vectors) <= len(label_vectors):
            cur = random.sample(all_labels, 1)
            if cur in label_vectors:
                continue
            non_true_vectors.append(cur[0])
        l_v = []
        for i in label_vectors:
            try:
                l_v.append(label_dict[i])
            except:
                non_true_vectors.pop()
        left = np.vstack((np.vstack(l_v), np.vstack([label_dict[i] for i in non_true_vectors])))
        labels = [1 for _ in label_vectors]+[0 for _ in non_true_vectors]
        right = np.vstack([vector]*len(label_vectors+non_true_vectors))
        for l, r, label in zip(left, right, labels):
            ret = (np.hstack((l, r)).reshape((1, 2000)), np.array([label]).reshape((1)))
            yield ret

wv = gensim.models.KeyedVectors.load("patents.wv", mmap='r')
porter = PorterStemmer()
ret = []
ret_labels = []
placeholder_vec = [0 for _ in range(1000)]
def get_sentence_vector(sentence):
    ret = []
    for word in sentence:
        try:
            ret.append(wv[porter.stem(word)])
        except:
            pass
    while len(ret) > 200:
        ret.pop(-1)
    while len(ret) < 200:
        ret.append(placeholder_vec)
    return np.vstack(ret)
def lstm_data_generator():
    with open('../dataset.csv', encoding="utf8") as csvfile:
        for row in csv.reader(csvfile):
            lstm_input_patent = get_sentence_vector(word_tokenize(row[1]))
            label_vectors = [convert_label_to_vector(z.strip().strip('\'\"')) for z in row[0].split(',')]
            non_true_vectors = []
            while len(non_true_vectors) <= len(label_vectors):
                cur = random.sample(all_labels, 1)
                if cur in label_vectors:
                    continue
                non_true_vectors.append(cur[0])
            l_v = []
            for i in label_vectors:
                try:
                    l_v.append(label_dict[i][1])
                except:
                    non_true_vectors.pop()
            non_trues = [label_dict[i][1] for i in non_true_vectors]
            left = l_v+non_trues
            labels = [1 for _ in label_vectors]+[0 for _ in non_true_vectors]
            for l, label in zip(left, labels):
                if np.any(np.isnan(lstm_input_patent)) or np.any(np.isnan(l)):
                    continue
                ret = ((lstm_input_patent, l), np.array([label]).reshape((1)))
                yield ret
```

```python
lstm_dataset = tf.data.Dataset.from_generator(lstm_data_generator, ((tf.float64, tf.float64), tf.float64), ((tf.TensorShape([200, 1000]), tf.TensorShape([100, 1000])), tf.TensorShape([1])))
lstm_dataset = lstm_dataset.batch(16, drop_remainder=True).repeat()
```

```python
import os
dataset = tf.data.Dataset.from_generator(data_generator, (tf.float64, tf.float64), (tf.TensorShape([1, 1003]), tf.TensorShape([1])))
test_dataset = tf.data.Dataset.from_generator(data_generator_test, (tf.float64, tf.float64), (tf.TensorShape([1, 1003]), tf.TensorShape([1])))
test_dataset = test_dataset.batch(128, drop_remainder=True).repeat()
train_dataset = dataset.batch(128, drop_remainder=True).repeat()
print(train_dataset)
print(test_dataset)
train_dataset.cache(filename='cached_dataset')
```

```python
#train_dataset = tf.data.Dataset.zip((tf.data.Dataset.from_tensors((training_vectors[:300000], training_labels[:300000])), tf.data.Dataset.from_tensors((training_vectors[300000:], training_labels[300000:]))))

#model = tf.keras.Sequential()
#model = tf.keras.models.Sequential([
#    tf.keras.layers.LSTM(200, input_shape=(200, 1000)),
#    tf.keras.layers.Dense(100, activation='relu'),
#    tf.keras.layers.Dense(50, activation='relu'),
#    tf.keras.layers.Dense(3)
#])

input_lstm = tf.keras.Input(shape=(200, 1000))
input_label_lstm = tf.keras.Input(shape=(100, 1000))

patent_lstm = tf.keras.layers.LSTM(50, input_shape=(200, 1000), activation = 'tanh', recurrent_activation = 'sigmoid', recurrent_dropout = 0, unroll=False, use_bias=True
)(input_lstm)
label_lstm = tf.keras.layers.LSTM(50, input_shape=(100, 1000), activation = 'tanh', recurrent_activation = 'sigmoid', recurrent_dropout = 0, unroll=False, use_bias=True
)(input_label_lstm)

concat = tf.keras.layers.Concatenate(axis=1)([patent_lstm, label_lstm])
dense_1 = tf.keras.layers.Dense(40, activation='relu')(concat)
output_binary = tf.keras.layers.Dense(1, activation="sigmoid")(dense_1)
model = tf.keras.Model(inputs=[input_lstm, input_label_lstm], outputs=[output_binary])

#saver = tf.train.Saver(max_to_keep=4, keep_checkpoint_every_n_hours=2)
model.summary()
model.compile(loss=['binary_crossentropy'],
              optimizer=tf.optimizers.Adam(1e-4),
              metrics=['accuracy'])
```
